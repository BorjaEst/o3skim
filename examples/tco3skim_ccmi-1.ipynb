Use `load_tco3` *function* inside `o3skim` *module* indicating the **path** for the model files and **model** name to load a normalized xarray dataset.

```python
import o3skim
ccmi_access = o3skim.load_tco3(
    path="tests/datasets_tco3/CCMI-1_ACCESS_ACCESS-CCM-refC2/*.nc", 
    model="CCMI-1_ACCESS"
)
```

```python
ccmi_access
```

Use `lon_mean` or `lat_mean` *functions* inside `o3skim` to calculate `Zonal Mean` or `Meridional Mean`.

```python
skimmed = o3skim.lon_mean(ccmi_access)
skimmed
```
